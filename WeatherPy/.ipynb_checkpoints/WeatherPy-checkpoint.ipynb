# WeatherPy

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress

# Import API key
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "../output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(lat_range[0], lat_range[1], size=1500)
lngs = np.random.uniform(lng_range[0], lng_range[1], size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls

```python
# Build query url
units = 'imperial'
query_url = f"http://api.openweathermap.org/data/2.5/weather?appid={weather_api_key}&units={units}&q="

# Create empty lists to hold location and weather data
cityNames = []
country = []
date = []
lat = []
lng = []
maxTemp = []
humidity = []
cloudiness = []
windSpeed = []

from pprint import pprint

# Perform a weather check on each city using a series of successive API calls.
# Include a print log of each city as it's being processed (with the city number and city name).
print('---------------------------------------------------')
print('Retrieving Data...')
print('---------------------------------------------------')

city_number = 0

for city in cities:
    city_number = city_number +1
    print(f"Processing City Number {city_number}: {city}")
    response = requests.get(query_url + city).json()
    
    # Try to grab and append the location and weather info for the city if available
    try:
        cityNames.append(response['name'])
        country.append(response['sys']['country'])
        date.append(response['dt'])
        lat.append(response['coord']['lat'])
        lng.append(response['coord']['lon'])
        maxTemp.append(response['main']['temp_max'])
        humidity.append(response['main']['humidity'])
        cloudiness.append(response['clouds']['all'])
        windSpeed.append(response['wind']['speed'])
        print(f"{city} found! Stats have been added.")
        print('------')
        # for testing:
        #pprint(response)
    
    # If city is unavailable, do the following...    
    except:
        print('City not found, skipping to next city...')
        print('------')
        pass
    
    # for testing:
    #if city_number == 10:
        #break
    
print('---------------------------------------------------')
print('Data Retrieval Complete')
print('---------------------------------------------------')
```

### Convert Raw Data to DataFrame


```python
# Create dataframe with location and weather data
weather_df = pd.DataFrame({'City': cityNames,
                          'Country': country,
                          'Date': date,
                          'Lat': lat,
                          'Lng': lng,
                          'Max_Temp': maxTemp,
                          'Humidity': humidity,
                          'Cloudiness': cloudiness,
                          'Wind_Speed': windSpeed})

# Display the dataframe
display(weather_df)

# Export dataframe into a .csv
weather_df.to_csv('../output_data/cities.csv', encoding='utf-8', index=True)
```

## Linear Regressions

```python
# Separate the data into northern hemisphere (Lat >= 0) and southern hemisphere (Lat < 0)
hem_N = weather_df.loc[weather_df['Lat'] >=0]
hem_S = weather_df.loc[weather_df['Lat'] <0]

# Create function to run linear regression on plots
def lin_reg(x, y):
    
    # Run the linear regression
    (slope, intercept, rvalue, pvalue, stderr) = linregress(x, y)
    
    # Set the regression line for plotting
    regressLine = x * slope + intercept
    
    # Create the regression equation string to display in plot
    regressEq = 'y = ' + str(round(slope,2)) + 'x + ' + str(round(intercept,2))
    
    # Print the results of the regression
    print(f'The correlation between the variables is {round(rvalue,2)}')
    print(f"The coefficient of determination is {round(rvalue**2,2)}")
    print(f"The equation of the regression line is {regressEq}")
    
    # Return the regression line and equation
    return(regressLine, regressEq)
```

### Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
# Run the linear regression
regressLine, regressEq = lin_reg(hem_N['Lat'], hem_N['Max_Temp'])

# Create the scatter plot with linear regression model
fig1 = plt.figure(figsize=(10,7))
plt.scatter(hem_N['Lat'], hem_N['Max_Temp'], marker='o', facecolors='xkcd:azure', edgecolors='black')
plt.plot(hem_N['Lat'], regressLine, 'r-')
plt.annotate(regressEq, (10,35), fontsize=15, color='red')

# Include proper labeling - title (with date of analysis) and axes labels
plt.title(f"Northern Hemisphere - Max Temp vs. Latitude ({day}/{month}/{year})")
plt.ylabel("Max Temperature (F)")
plt.xlabel("Latitude")
plt.grid(True)

# Save figure to .png
plt.savefig('../output_data/NH_TempVsLat.png')

# Display plot
plt.show()
```

The above plot shows a strong negative relationship between maximum temperature and latitude in the northern hemisphere. In other words, the further north the city is from the equator (or Lat = 0), the lower their maximum temperature is.
